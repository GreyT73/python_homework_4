```python
houses_waterfront = houses['waterfront'].value_counts()

houses_waterfront
```

```python
plt.pie(houses_waterfront, autopct='%1.1f%%')
plt.legend(houses_waterfront.index)
plt.title("Распределение в зависимости от вида на набережную");
```

```python
houses_floors = houses['floors'].value_counts()
houses_floors
```

```python
plt.pie(houses_floors, autopct='%1.1f%%')
plt.legend(houses_floors.index)
plt.title("Распределение в зависимости от количества этажей");
```

```python
houses_condition = houses['condition'].value_counts()
plt.pie(houses_condition, autopct='%1.1f%%')
plt.legend(houses_condition.index)
plt.title("Распределение в зависимости от состояния");
```

Домов с видом на набережную очень мало - менее одного процента. Половина домов - дома с одним этажом. Большинство домов имеют среднее состояние, около четверти - хорошее. В плохом состояние домов почти нет.

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
corr_matrix = houses.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix)
```

```python
sns.jointplot(x=houses['sqft_living'], y = houses['price'], kind = 'reg')
plt.title("Завсисимость стоимости от площади")
plt.xlabel("Площадь")
plt.ylabel("Стоимость")
plt.ticklabel_format(style='plain');
```

Площадь является одним из самых значимых фактором, влияющим на стоимость жилья. Корреляция наиболее близка к 1.

```python
sns.boxplot(x=houses['grade'], y = houses['price'], whis = 5)
plt.title("Зависимость стоимости от состояния")
plt.xlabel("Состояние")
plt.ylabel("Стоимость");
```

Состояние также сильно влияет на стоимость - чем лучше состоягие, тем выше стоимость

```python
sns.scatterplot(x= houses['floors'], y = houses['price'])
plt.title("Зависимость стоимости от этажей")
plt.xlabel("Этажей")
plt.ylabel("Стоимость");
```
